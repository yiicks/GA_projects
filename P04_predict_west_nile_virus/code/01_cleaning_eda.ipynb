#### Functions

````python
def check_missing_data(column):
    return weather[weather[column] == "M"]
````

````python
def impute_missing_data(column):
    for k, v in enumerate(weather[column]):
        if v == "M" and weather.iloc[k]["Station"] == 1:
            weather.loc[k, column] = weather[column][k + 1]
        if v == "M" and weather.iloc[k]["Station"] == 2:
            weather.loc[k, column] = weather[column][k - 1]
    weather[column] = weather[column].astype(int)
````

````python
def impute_missing_float_data(column):
    for k, v in enumerate(weather[column]):
        if v == "M" and weather.iloc[k]["Station"] == 1:
            weather.loc[k, column] = weather[column][k + 1]
        if v == "M" and weather.iloc[k]["Station"] == 2:
            weather.loc[k, column] = weather[column][k - 1]
    weather[column] = weather[column].astype(float)
````

````python
def to_celsius(temp):
    celsius = ((temp - 32.0) * 5.0) / 9.0
    return float(celsius)
````

````python
def humidity(row):
    row["Humidity"] = round(
        100
        * (
            np.exp((17.625 * row["DewPoint"]) / (243.04 + row["DewPoint"]))
            / np.exp((17.625 * row["Tavg"]) / (243.04 + row["Tavg"]))
        )
    )
    return row
````

````python
def daylength(time):
    min = (time // 100) * 60
    min += time % 100
    return min
````

````python
def cleaner(df):
    # change title to lower case
    df.columns = df.columns.str.lower()

    # drop unnecessary columns
    try:
        df.drop(
            columns=[
                "addressaccuracy",
                "addressnumberandstreet",
                "address",
                "block",
                "street",
            ],
            inplace=True,
        )
    except KeyError:
        pass

    # convert date time
    df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")

    # create year, month/week in dataset
    df["year"] = (df["date"].dt.strftime("%Y")).astype(int)
    df["weekofyear"] = df.date.dt.isocalendar().week

    # total mozzies as integer
    try:
        df["nummosquitos"] = df["nummosquitos"].astype("int64")
    except KeyError:
        pass

    # display final datatypes
    display(df.dtypes)
````

````python
def dum_wkyr(dataset):

    # dummify weekofyear
    dataset = pd.get_dummies(dataset, columns=["weekofyear"], prefix="dumweek")

    # only weeks 28-39 are significant
    # drop all other dummified weeks

    try:
        dataset.drop(
            columns=[
                "dumweek_22",
                "dumweek_23",
                "dumweek_24",
                "dumweek_25",
                "dumweek_26",
                "dumweek_27",
                "dumweek_40",
                "dumweek_41",
            ],
            inplace=True,
        )
    except KeyError:
        dataset.drop(
            columns=[
                "dumweek_22",
                "dumweek_23",
                "dumweek_24",
                "dumweek_25",
                "dumweek_26",
                "dumweek_27",
                "dumweek_40",
            ],
            inplace=True,
        )
    display(dataset.head())
    return dataset
````

Impute missing data using suntime for sunrise and sunset timing based on latitude and longitude of the weather station 2, since only weather station 2 has missing values.

````python
latitude = 41.786
longitude = -87.752

sun = Sun(latitude, longitude)

central = timezone("US/Central")

sunrise_list = []
sunset_list = []
for value in weather["Date"]:
    abd = value
    abd_sr = sun.get_local_sunrise_time(value, local_time_zone=central)
    abd_sr = abd_sr - abd_sr.dst()
    abd_ss = sun.get_local_sunset_time(value, local_time_zone=central)
    abd_ss = abd_ss - abd_ss.dst()
    sunrise_list.append(abd_sr.strftime("%H%M"))
    sunset_list.append(abd_ss.strftime("%H%M"))
````

````python
index = weather.loc[weather["Station"] == 2].index

for i in index:
    weather.loc[i, "Sunrise"] = sunrise_list[i]
    weather.loc[i, "Sunset"] = sunset_list[i]
````

Colder temperatures limit the reproduction of Culex mosquitoes and, along with shortening daylength, can induce some species to enter reproductive diapause. In this section, we first examine the distribution of mosquito species against them being carriers of the WNV.

Only 3 species from the train data set seem to be carriers of the WNV: 
* Culex Pipens/Restuans
* Culex Restuans
* Culex Pipiens

We then decided to group these in 2 main species groups -  1 for both culex pipiens/restuans and culex pipiens, 2 for culex restuans as we found that culex pipiens/restuans and culex pipiens have similar likelihoods of being WNV carriers.  

````python
# show total mozzies by species

plt.figure(figsize=(10, 7))
plot = sns.barplot(
    x="species", y="nummosquitos", hue="wnvpresent", data=train, estimator=sum
)
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Total Number of Mozzies", fontsize=14)
plt.xlabel("Species", fontsize=14)
plt.legend(loc="upper right", title="Virus")
plt.title("Mozzies by Species (2007-2013)", fontsize=20)
plt.show()
````

````python
# Percentage of species with WNV

wnv_species = ["CULEX PIPIENS/RESTUANS", "CULEX PIPIENS", "CULEX RESTUANS"]

for i in wnv_species:
    percent = round(
        (
            (
                train[(train["wnvpresent"] == 1) & (train["species"] == str(i))][
                    "nummosquitos"
                ].sum()
            )
            / (train[train["species"] == str(i)]["nummosquitos"].sum())
            * 100
        ),
        2,
    )
    print(f"Percentage of {i} with west nile virus is: {percent}%")
````

### Spray vs mosquito population and WNV

In this section, we explore the effect of spraying on the total mosquito population, and whether WNV was present. 

As we only have 2 years of spray data (2011 and 2013), we only plotted the charts for these 2 years.

#### Spraying in 2011

Spraying was only done twice, with little to no effect on WNV.

For the first spray, total mosquito population decreased, but WNV was constant. 

After the 2nd spray, both mosquito population and WNV mosquitoes went up, suggesting that 
1) the sprays are not effective, or 

2) there could be other factors that increased mosquito breeding / propagation of WNV. 

````python
# when were sprays implemented

# Split the dataframe based on year
spray_df_2013 = spray[spray["date"].dt.year == 2013]
spray_df_2011 = spray[spray["date"].dt.year == 2011]

spray_2011 = spray_df_2011["date"].unique()[:2]
spray_2011

spray_2013 = spray_df_2013["date"].unique()[:8]
spray_2013
````

````python
moz_2011 = train[train["date"].dt.year == 2011].groupby("date")["nummosquitos"].sum()
wnv_2011 = train[train["date"].dt.year == 2011].groupby("date")["wnvpresent"].sum()


fig, ax = plt.subplots(nrows=2, figsize=(15, 6))
ax[0].plot(moz_2011.index, moz_2011.values, color="black")
for date in spray_2011:
    ax[0].axvline(x=date, color="green", ls="-")
ax[0].set_title("total mosquitos 2011")
ax[0].legend(["mosquitos", "spray"], loc="upper right")


ax[1].plot(wnv_2011.index, wnv_2011.values, color="blue")
for date in spray_2011:
    ax[1].axvline(x=date, color="green", ls="-")
ax[1].set_title("wnv 2011")
ax[1].legend(["wnv", "spray"], loc="upper right")


plt.tight_layout()
````

#### Spraying in 2013 

For 2013, mosquitoes were sprayed mosquitoes extensively. It seemed to work in the beginning.

However in mid-August, despite spraying the mosquito population (that had WNV) went up although total number of mosquitoes kind of remained constant. Could it be due to weather factors? 

````python
moz_2013 = train[train["date"].dt.year == 2013].groupby("date")["nummosquitos"].sum()
wnv_2013 = train[train["date"].dt.year == 2013].groupby("date")["wnvpresent"].sum()


fig, ax = plt.subplots(nrows=2, figsize=(15, 6))
ax[0].plot(moz_2013.index, moz_2013.values, color="black")
for date in spray_2013:
    ax[0].axvline(x=date, color="green", ls="-")
ax[0].set_title("total mosquitos 2013")
ax[0].legend(["mosquitos", "spray"], loc="upper right")


ax[1].plot(wnv_2013.index, wnv_2013.values, color="blue")
for date in spray_2013:
    ax[1].axvline(x=date, color="green", ls="-")
ax[1].set_title("wnv 2013")
ax[1].legend(["wnv", "spray"], loc="upper right")


plt.tight_layout()
````

### Mozzies vs Weather

Observations with interactions from weather columns: 

1) Likelihood of WNV is higher for culex pipiens+mixed breed with lower amount of rainfall. Similar results with lag, the optimal rainall for WNV seems to be 0-0.5mm for propagation of WNV. 

2) Humidity between 55-80 seems like the best conditions for WNV in species 1

3) Regardless of temperature, culex pipiens+mixed breed still seem to propogate WNV, effects of temperature are more prominent for culex restuans (>25 deg more likely for WNV) 

4) Optimal pressure for wnv propogation - between 29-29.5 

5) Daylength and Resultant speed seem like pretty useless features - mosquitos propogate WNV regardless - esp for pipiens+mixed breed

6) WNV more prevalent in weeks 28-37 (but why?)
